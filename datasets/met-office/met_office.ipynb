# Met office (UK) deterministic forecast models

A light notebook demonstrating access of the met office STAC items in our test database.

```python
from IPython.display import display
from pystac_client import Client

client = Client.open("https://planetarycomputer-test.microsoft.com/stac")
collection_search = client.collection_search(q="met-office")
collections = list(collection_search.collections())
for collection in collections:
    print(collection.id)
```

```python
for collection in collections:
    item_search = client.search(collections=[collection.id])
    items = list(item_search.items())
    if items:
        print(f"Found {len(items)} items for collection: {collection.id}")
        display(items[0])
    else:
        print("No items found for collection:", collection.id)
```
