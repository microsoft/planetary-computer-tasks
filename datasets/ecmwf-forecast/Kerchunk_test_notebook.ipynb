```python
#Note: Change the item path to point to a STAC item you recently created.

#read in stac item
import pystac
item = "https://pct-apis-staging.westeurope.cloudapp.azure.com/stac/collections/ecmwf-forecast-wave-fc/items/ecmwf-2023-10-19T00-wave-fc-0h"
item = pystac.read_file(item)
```

```python
#stac item with kerchunk indices
import fsspec
import xarray as xr
import stactools.ecmwf_forecast

m = fsspec.get_mapper('reference://', fo=item.assets['data'].to_dict()['kerchunk:indices'])
ds = xr.open_dataset(m, engine='zarr', consolidated=False)
ds['swh'].mean()
```

```python
#original .grib2 file
import urllib.request

file = 'https://ai4edataeuwest.blob.core.windows.net/ecmwf/20231019/00z/0p4-beta/wave/20231019000000-0h-wave-fc.grib2'
filename, _ = urllib.request.urlretrieve(file)
ds = xr.open_dataset(filename, engine="cfgrib")
ds['swh'].mean()
```
